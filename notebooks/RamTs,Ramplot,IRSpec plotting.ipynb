```python


datpath = "/cluster/home/schmidtn/master/wahlprak/charmm/apa-/gas/extended/gas-extended/"


def drawRamTs():

    dat_files = list(Path(datpath).glob("*dat"))
    RamTs = [_ for _ in dat_files if "ramts" in str(_) ]
    RamTsAr = np.genfromtxt(RamTs[0])
    #print(RamTsAr)
    Ts = RamTsAr[:,0]
    Phi = RamTsAr[:,1]
    Psi = RamTsAr[:,2]

    plt.plot(Ts,Phi,linestyle="", marker="o", c="k", markersize=0.01)
    plt.plot(Ts,Psi, linestyle="", marker="o", c="r", markersize=0.01)
    plt.savefig("tst.pdf", bbox_inches="tight")

    print("done")





drawRamTs()




    
```
